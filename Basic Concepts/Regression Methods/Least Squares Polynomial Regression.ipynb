```python
import numpy as np
from scipy.linalg import lstsq, norm, eig, pinv
from numpy.random import rand, randn, permutation
import scipy.linalg
import matplotlib.pyplot as plt
```

```python
# Fit a LS polynomial approximation to clean data
t = np.linspace(0,2,16)
y = 0.5*np.exp(0.8*t)

polyOrder1 = 15 # up to x**15
polyOrder2 = 2  # up to x**2

A1 = np.zeros((len(t),polyOrder1+1))
A2 = np.zeros((len(t),polyOrder2+1))

for i in range(polyOrder1+1):
    A1[:,i] = t**i

x1 = lstsq(A1,y)[0]
p1 = A1@x1

for i in range(polyOrder2+1):
    A2[:,i] = t**i
    
x2 = lstsq(A2,y)[0]
p2 = A2@x2

plt.plot(t,y,'b+',label='y(t)')
plt.plot(t,p1,label='poly. order = 15')
plt.plot(t,p2,label='poly. order = 2')
plt.legend()
plt.grid();
```

### Clean data (above):###
**order 2 and 15 polynomials fit the data well**

### Noisy data (below): ###
**order 15 overfits, while order 2 gets an ok approximation**


```python
# Now fit to noisy data
y = y + np.random.rand(len(t))

for i in range(polyOrder1+1):
    A1[:,i] = t**i

x1 = lstsq(A1,y)[0]
p1 = A1@x1

for i in range(polyOrder2+1):
    A2[:,i] = t**i
    
x2 = lstsq(A2,y)[0]
p2 = A2@x2


plt.plot(t,y,'b+',label='y(t)')
plt.plot(t,p1,label='p15')
plt.plot(t,p2,label='p2')
plt.legend()
plt.grid();
```
