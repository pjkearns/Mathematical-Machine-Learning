```python
"""
Authors: Phillip Kearns, John Lipor
KRR demo - syntheticTest
python version: 3.7.2
"""
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import randn, rand
from numpy import sin
from KRR import KRR
```

```python
## generate and plot triaining data
N = 50
X = np.linspace(0, 1, N).reshape(N,1)
f1 = 30
f2 = 15
y = 3.0 * np.vstack((sin(f1 * X[:int(N/2)]), sin(f2 * X[int(N/2):]))) - 2*X + 0.1 * randn(N, 1)

plt.figure()
plt.scatter(X, y, marker='x',label='Sampled Function');
```

```python
### Task 2: KRR on training data: 
lambda_ = 0.1   # vary this parameter
sigma = 0.1     # vary this parameter

fig,axes = plt.subplots(nrows=4, ncols=1,figsize=(5,10))

for ii, sigma in enumerate([0.01, 0.05, 0.1, 0.5]):
    ytrain, _ = KRR(X, y, lambda_, sigma)
    axes[ii].scatter(X, y, marker='x',label='Sampled Function');
    axes[ii].plot(X, ytrain.T, color='C1', label=r'Predicted,$\sigma = $%.2f'%sigma)
    axes[ii].legend(loc='upper right',bbox_to_anchor=(1.5,1.0));
```

```python
### Task 3: KRR on test data:
Ntest = 10
Xtest = rand(Ntest, 1)

fig,axes = plt.subplots(nrows=4, ncols=1,figsize=(5,10))

for ii, sigma in enumerate([0.01, 0.05, 0.1, 0.5]):
    ytrain, ytest = KRR(X, y, lambda_, sigma, Xtest=Xtest)
    axes[ii].scatter(X, y, marker='x',label='Sampled Function');
    axes[ii].plot(X, ytrain.T, color='C1', label=r'Train Pred,$\sigma = $%.2f'%sigma)
    axes[ii].scatter(Xtest, ytest.T, marker='o', c='C2', label=r'Test, Pred,$\sigma = $%.2f'%sigma);
    axes[ii].legend(loc='upper right',bbox_to_anchor=(1.5,1.0));
```
