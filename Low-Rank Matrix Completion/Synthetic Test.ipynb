```python
"""
Author: John Lipor
LMRC demo - syntheticTest
python version: 3.7.3
"""
import numpy as np
import time
import matplotlib.pyplot as plt
from numpy.random import randn, rand
from numpy.linalg import norm 
from scipy.io import loadmat
from sIRLSp import sIRLSp
```

```python
M = 50
N = 50
r = 5

## true data matrix
Xtrue = randn(M, r) @ randn(r, N)

## create mask to sample samProb percent of entries
sampProb = 0.4
mask = rand(M, N) < sampProb 
X = Xtrue * mask

## run you algorithm
p = 1
start = time.time()
Xhat = sIRLSp(X, mask, p)
err = norm(Xhat - Xtrue, ord='fro')/norm(Xtrue, ord='fro')
print(f"Estimated error: {err}")
print(f"Time: {time.time() - start}")

plt.subplot(131)
plt.imshow(Xtrue)
plt.title('Original')
plt.subplot(132)
plt.imshow(X)
plt.title('Input')
plt.subplot(133)
plt.imshow(np.real(Xhat))
plt.title('Completed');
```
